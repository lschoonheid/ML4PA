## Exercise 1 - A toy strong lensing image

Our "simulator" produces images with one or more rings, where the radius, the width and the number of rings can be modified.  Think of this model as a very simple toy model for producing for instance images of graviationally lensed systems.  The radius would then correspond to the "Einstein radius" of the gravitational lensing system and tell us something about the mass of the lensing galaxy. The width would correspond to the size of the lensed source galaxy.  Normally, you would not get multiple rings in one image, but in our toy model we do, to make it a bit more difficult for the networks to learn the parameters.

A real strong lensing image looks like this:

lens.jpeg


```python
def image_model(radius = 0.5, sigma = 0.1, width = 0.15, n = 1):
    x = np.linspace(-2, 2, 32)
    X, Y = np.meshgrid(x, x)
    
    imgs = []
    #Generate n rings
    for i in range(n):
      #Random centering of the ring
      x0, y0 = np.random.uniform(-1, 1, 2)
        
      #Change meshgrid values that come across with the ring in
      R = ((X-x0)**2 + (Y-y0)**2)**0.5
      img = np.exp(-(R-radius)**2/width**2/2)
      img = img

      imgs.append(img)
    
    # Standard variance and zero mean
    img = sum(imgs)

    # Add image noise
    img += np.random.randn(*img.shape)*sigma

    # Remove mean and variance, which makes it easier to train for the NN
    img -= img.mean()
    img /= img.std()
    return img
```

### Parameter priors

In later exercises, we will create images for a range of parameters.  Those parameters are randomly drawn in the function below, with the radius ranging from 0.5 to 1.5 and the width from 0.1 to 1.0.

```python
def draw_from_prior():
    # returns random draw from p(z)
    radius = np.random.uniform(0.5, 1.5)
    width = np.random.uniform(0.1, 1.0)
    return {'radius': radius, 'width': width}
```

## Exercise 2 - Parameter regression

Our goal in this exercise is to train a network that estimates the radius of the rings in the images that we have generated in the previous exercise. More speficically, we will obtain a function that uses such an image as input, and predicts the radius of the rings in that image:

radius = NN(image).



### Convolutional neural network 

We define here a standard convolutional neural network with two convolutional layers and three dense layers at the bottom.  The forward function takes as argument an image ($x$), and returns a single number (the radius).


```python
class RegressionCNN(nn.Module):
    def __init__(self):
        super(RegressionCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 5, 3)
        self.conv2 = nn.Conv2d(5, 10, 3)
        self.fc1 = nn.Linear(40, 1024)
        self.fc2 = nn.Linear(1024, 128)
        self.fc3 = nn.Linear(128, 1)
        self.pool = nn.MaxPool2d(3)

    def forward(self, x):
        
        #when forward is called, x is an image
        x = x.unsqueeze(0).unsqueeze(0)
        x = self.conv1(x)
        x = self.pool(x)
        x = self.conv2(x)
        x = self.pool(x)
        x = x.flatten() 

        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)
        #after running through the neural network, x is a single number predicting the ring radius

        return x
```

### The training algorithm

In the following cell, we train the neural network by looping through the input data (many (image, radius) pairs) and optimizing the neural network `net`, such that the predicted value matches the actual ring radius.

```python
# Initialize network
net = RegressionCNN()
optimizer = optim.Adam(net.parameters(), lr = 1e-3)
loss_history = []

# Training network
net.train()  # Set network to training mode
Niter = 16000  # Number of training iterations

for i in tqdm(range(Niter)):
    optimizer.zero_grad()  # Reset internal gradients

    radius_true = draw_from_prior()['radius'] # Pick a random radius 
    img = image_model(radius = radius_true, n = 1)

    img = torch.tensor(img).float()
    
    radius_estimated = net(img)
    
    # Training minimizes the loss function
    loss = (radius_estimated - radius_true)**2
    # print(radius_estimated, radius_true, loss)
    # loss_history.append(loss.detach().numpy())

    loss.backward()
    optimizer.step()

plt.plot(loss_history)
plt.xlabel('Training iterations')
plt.ylabel('Training loss')
plt.show()
```

### Testing the reconstruction accuracy

We now can compare the predictions of the network with the ground truth.

## Exercise 3 - Classification

Above, we were performing parameter estimation. Now we consider a simple classification problem.  The goal is to train a network that predicts whether a specific image has 4 or 5 rings.

### Generative model

The model now just takes a class label (0 or 1), and returns images with either 4 or 5 rings.

```python
def model_clas(c):
    n = 4 + c
    img = image_model(radius = 0.5, n = n, width = 0.1)
    return img

img = model_clas(1)
plt.imshow(img)
```

### Classification network

The network here has the same structure as the one used above for regression. The only difference is that at the end we map the output parameter onto the interval (0, 1).  This will allow us to interpret the output $p = NN(img)$ as the probability that the input image corresponds to class 0.  The probability that the image belongs to class 1 is then simply $1-p$.


```python
class ClassificationCNN(nn.Module):
    def __init__(self):
        super(ClassificationCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 5, 3)
        self.conv2 = nn.Conv2d(5, 10, 3)
        self.fc1 = nn.Linear(40, 1024)
        self.fc2 = nn.Linear(1024, 128)
        self.fc3 = nn.Linear(128, 1)
        self.pool = nn.MaxPool2d(3)

    def forward(self, x):
        x = x.unsqueeze(0).unsqueeze(0)
        x = self.conv1(x)
        x = self.pool(x)
        x = self.conv2(x)
        x = self.pool(x)
        x = x.flatten() 

        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)

        x = torch.sigmoid(x)  # We use sigmoid to map onto (0, 1)

        return x
```

```python
# Initialize network
netC = ClassificationCNN()
optimizer = optim.Adam(netC.parameters(), lr = 1e-3)
loss_history = []

# Training network
netC.train()  # Set network to training mode
Niter = 16000  # Number of training iterations

for i in tqdm(range(Niter)):
    optimizer.zero_grad()  # Reset internal gradients

    img0 = model_clas(0)  # Image with 4 rings
    img1 = model_clas(1)  # Image with 5 rings

    img0 = torch.tensor(img0).float()
    img1 = torch.tensor(img1).float()
    
    p0 = netC(img0)
    p1 = netC(img1)

    # Binary cross-entropy loss function
    loss = -torch.log(p0) - torch.log(1-p1)
    
    # Training minimizes the loss function
    loss_history.append(loss)

    loss.backward()
    optimizer.step()

# plt.plot(loss_history)
plt.xlabel('Training iterations')
plt.ylabel('Training loss')
plt.show()
```

### Plotting of validation results

```python
img0 = model_clas(1)  # Image with 4 rings

plt.imshow(img0)

img0 = torch.tensor(img0).float()
p0 = netC(img0).detach().numpy()[0]

print("Probability 4 rings: %.2f"%p0)
print("Probability 5 rings: %.2f"%(1-p0))
```

## Exercise 4 - Posterior estimation

Up to now, we considered a regression problem and a classification problem.  In the case of the regression problem, we only got a single number estimator for the radius, without any information about the uncertainty of that prediction.

However, for the classification problem, we actually obtained probabilities, and could see that these probabilities (or uncertainties of the classification) can drastically change depending on how clearly the number of rings in a specific image can be recognized.

Here, we will try to bring both ideas together, and train a network that gives us distributions for parameters, so that we have a measure for how confident we are with our prediction. To do this, some concepts from probability theory will be used that you may not be familiar with. At the end of this notebook this will hopefully become more clear.

### Likelihood-to-evidence ratio estimation

As discussed in the lecture, the main idea of likelihood-to-evidence ratio estimation is to train a classification network that can discriminate two hypothesis (or two "classes"):

$H_0$: The (image, radius) pair belong together

$H_1$: The (image, radius) pair are random.

Technically, we can generate examples for $H_0$ by first generating a random radius, and then the corresponding image.


```python
radius = draw_from_prior()["radius"]
img = image_model(radius = radius, n = 2)
plt.imshow(img)
plt.title("Radius: %.3f"%radius)
```

We can generate examples for $H_1$ by first generating a random image (with some random radius), and pair this up with another unrelated radius

```python
radius = draw_from_prior()["radius"]
img = image_model(radius = radius, n = 2)
radius = draw_from_prior()["radius"]

plt.imshow(img)
plt.title("Radius: %.3f"%radius)
```

```python
class RatioCNN(nn.Module):
    def __init__(self):
        super(RatioCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 5, 3)
        self.conv2 = nn.Conv2d(5, 10, 3)
        self.fc1 = nn.Linear(41, 1024)
        self.fc2 = nn.Linear(1024, 128)
        self.fc3 = nn.Linear(128, 1)
        self.pool = nn.MaxPool2d(3)
    
    #Here we forward both x and 
    def forward(self, x, z):
        x = x.unsqueeze(0).unsqueeze(0)
        x = self.conv1(x)
        x = self.pool(x)
        x = self.conv2(x)
        x = self.pool(x)
        x = x.flatten() 

        x = torch.cat([x, z.unsqueeze(0)])

        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)

        x = torch.sigmoid(x)  # We use sigmoid to map onto (0, 1)

        return x
```

```python
# Initialize network
netR = RatioCNN()
optimizer = optim.Adam(netR.parameters(), lr = 1e-3)
loss_history = []

# Training network
netR.train()  # Set network to training mode
Niter = 10000  # Number of training iterations

for i in tqdm(range(Niter)):
    optimizer.zero_grad()  # Reset internal gradients

    r_true  = draw_from_prior()['radius']
    r_false = draw_from_prior()['radius']

    img = image_model(radius = r_true, n = 3)

    img = torch.tensor(img).float()
    r_true = torch.tensor(r_true)
    r_false = torch.tensor(r_false)
    
    d_true = netR(img, r_true)
    d_false = netR(img, r_false)

    # Binary cross-entropy loss function
    loss = -torch.log(d_true) - torch.log(1-d_false)
    
    # Training minimizes the loss function
    loss_history.append(loss)

    loss.backward()
    optimizer.step()

# plt.plot(loss_history)
plt.xlabel('Training iterations')
plt.ylabel('Training loss')
plt.show()
```

### Posterior distribution

We create one image with a fixed radius `r_true` and generate a list of 30 radii in the range of 0.5-1.5. Next we let the RatioNN model predict the probability that these radii are correct for this image. Plotting these probabilities, we get a distribution of how probable it is to find a specific radius, judging on our model. This is called the posterior distribution.

```python
r_true  = draw_from_prior()['radius']
img = image_model(radius = r_true, n = 3)

plt.figure(figsize = (10, 5))
plt.subplot(121)
plt.imshow(img)

img = torch.tensor(img).float()
r_true = torch.tensor(r_true)

rlist = np.linspace(0.5, 1.5, 30)
dlist = []
for r in rlist:
    r = torch.tensor(r).float()
    d = netR(img, r).detach().numpy()[0]
    dlist.append(d)

plt.subplot(122)
plt.plot(rlist, dlist, '--',color='lightgreen', label = 'Posterior Probability distribution')
plt.axvline(r_true, label = 'Actual Radius')
plt.ylim(0,1)
plt.xlim(0,2)
plt.xlabel('Radius')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

# Extra example of posterior distribution and Bayes' Theorem.

In this notebook, we tried to make predictions on data with a mathematical model called a Neural Network. This machine learning algorithm or model is a specific way of thinking about the structured relationships/interactions in the data. In this way, a model can be thought of as a hypothesis about the relationships in the data, such as the relationship between input (X) and output (y). In other words, applying machine learning is nothing more than the testing and analysis of different hypotheses (models) on a given dataset. 


Bayes' Theorem provides a probabilistic model to describe the relationship between data (D) and a hypothesis (h):

$P(h|D) = \frac{P(D|h)  P(h)}{P(D)}$.

Breaking this down, it says that the probability of a given hypothesis holding/being true given some observed data can be calculated as follows:

- We take the probability of observing the data given the hypothesis 
- Multiply that by the probability of the hypothesis being true regardless of the data
- Divide that by the probability of observing the data regardless of the hypothesis.

Under this framework, each piece of the calculation has a specific name; for example:

- P(h|D): Posterior probability of the hypothesis (the thing we want to calculate).
- P(h): Prior probability of the hypothesis.

This P(h|D) is thus what we have computed in the previous cell (right figure). It is the probability that our hypothesis that the circles have a specific radius is correct, based on the judgment of our Neural Network.
